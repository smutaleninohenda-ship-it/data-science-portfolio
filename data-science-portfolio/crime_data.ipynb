```python
import pandas as pd
import numpy as np
from datetime import datetime
import random

# Set seed for reproducibility
np.random.seed(42)
random.seed(42)

# 14 regions of Namibia
regions = [
    'Khomas', 'Erongo', 'Oshana', 'Ohangwena', 'Oshikoto', 'Omusati',
    'Kunene', 'Otjozondjupa', 'Kavango East', 'Kavango West', 'Zambezi',
    'Hardap', 'ǁKaras', 'Omaheke'
]

crime_types = [
    'Theft', 'Assault', 'Robbery', 'Sexual Offence', 
    'Housebreaking', 'Fraud', 'Car Theft', 'Drug Related'
]

urban_regions = ['Khomas', 'Erongo', 'Oshana', 'Kavango East', 'Ohangwena']

start_date = datetime(2020, 1, 1)
end_date = datetime(2025, 5, 1)
dates = pd.date_range(start=start_date, end=end_date, freq='MS')

data = []

for date in dates:
    year = date.year
    month = date.month
    for region in regions:
        base_rate = 120 if region in urban_regions else 45
        if month == 12:
            seasonal_factor = 1.4
        elif month in [1, 11]:
            seasonal_factor = 1.1
        else:
            seasonal_factor = 1.0
        year_factor = 1 + (year - 2020) * 0.03
        for crime in crime_types:
            crime_factor = {
                'Theft': 1.2,
                'Assault': 0.9,
                'Robbery': 0.5,
                'Sexual Offence': 0.3,
                'Housebreaking': 0.8,
                'Fraud': 0.4,
                'Car Theft': 0.2,
                'Drug Related': 0.6
            }[crime]
            expected = base_rate * crime_factor * seasonal_factor * year_factor
            count = np.random.poisson(expected)
            count = max(0, int(count))
            data.append({
                'Year': year,
                'Month': month,
                'Date': date.strftime('%Y-%m-%d'),
                'Region': region,
                'CrimeType': crime,
                'CrimeCount': count
            })

df = pd.DataFrame(data)
df.to_csv('namibia_crime_data.csv', index=False)
print(f"Dataset generated with {len(df):,} rows.")
print(df.head())
```
